```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Mall_Customers.csv')
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s = 10, c = 'b')
plt.title("Mall Customers Data")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 50, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 50, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'black', label = 'Centroids')
plt.title("K-means Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

```python
X_test = np.array([[70, 80]]) # example new data point with annual income of 70k and spending score of 80
y_pred = kmeans.predict(X_test)
print('Predicted cluster label:', y_pred)
```

# Explanation
This code demonstrates how to use K-means clustering to group similar customers based on their Annual Income and Spending Score. We start by loading the Mall Customers dataset using Pandas, and selecting only two features for clustering: Annual Income and Spending Score.

Next, we create a scatter plot to visualize the data points of the two features selected. The scatter plot is displayed using Matplotlib with blue color.

Then, the K-means clustering algorithm is applied to the dataset with 4 clusters, using the KMeans function from the scikit-learn library. The n_clusters parameter is set to 4, max_iter to 300, n_init to 10, and init to 'k-means++'. A random seed of 0 is used for reproducibility.

After clustering, the data points are plotted again with each point's color indicating its cluster label. We use plt.scatter() function to plot each cluster with different color and label. Additionally, we have marked the centroids of each cluster as a black square, which is done using kmeans.cluster_centers_ and plt.scatter() function.

For testing, a new data point of [120, 10] is created and stored in the X_test variable. We then use the kmeans.predict() function to predict which cluster this point belongs to. The predicted cluster label is printed to the console.

Overall, this code demonstrates the process of using K-means clustering to group similar customers based on their Annual Income and Spending Score.
